```python
import pandas as pd

# 1. Criando um conjunto de dados (Simulando uma lista de vendas)
dados = {
    'Produto': ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Mouse', 'Monitor'],
    'Preco': [4500, 150, 300, 1200, 150, 1200],
    'Quantidade': [2, 10, 5, 3, 8, 4],
    'Loja': ['Centro', 'Sul', 'Norte', 'Centro', 'Norte', 'Sul']
}

# 2. Transformando em um DataFrame (Tabela do Pandas)
df = pd.DataFrame(dados)

# 3. Criando uma nova coluna: Faturamento Total por linha
df['Faturamento'] = df['Preco'] * df['Quantidade']

# 4. Análise: Qual o faturamento total por loja?
faturamento_por_loja = df.groupby('Loja')['Faturamento'].sum()

print("--- Tabela de Dados ---")
print(df)
print("\n--- Faturamento por Loja ---")
print(faturamento_por_loja)
```

```python
import pandas as pd

# 1. Criando um conjunto de dados (Simulando uma lista de vendas)
dados = {
    'Produto': ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Mouse', 'Monitor'],
    'Preco': [4500, 150, 300, 1200, 150, 1200],
    'Quantidade': [2, 10, 5, 3, 8, 4],
    'Loja': ['Centro', 'Sul', 'Norte', 'Centro', 'Norte', 'Sul']
}

# 2. Transformando em um DataFrame (Tabela do Pandas)
df = pd.DataFrame(dados)

# 3. Criando uma nova coluna: Faturamento Total por linha
df['Faturamento'] = df['Preco'] * df['Quantidade']

# 4. Análise: Qual o faturamento total por loja?
faturamento_por_loja = df.groupby('Produto')['Faturamento'].sum()

print("--- Tabela de Dados ---")
print(df)
print("\n--- Faturamento por Produto ---")
print(faturamento_por_loja)
```
