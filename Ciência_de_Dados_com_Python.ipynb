Escreva um programa que receba a idade de uma pessoa e diga se ela é:

adolescente (13–17)

adulto (18–59)

idoso (60+)

```python
idade = int(input('Digite sua idade: '))

if idade <= 12:
    print('Você é uma criança!')
elif idade <= 17:
    print('Você é adolescente!')
elif idade <= 59:
    print('Você é adulto!')
else:
    print('Você é idoso!')
```

```python
# dados de entrada
peso = float(input())
preco_por_tonelada = float(input())
tipo_cliente = input()

# Calculo o valor total sem desconto
valor_total = peso * preco_por_tonelada

# TODO Aplique o desconto conforme o tipo de cliente
if tipo_cliente == "Novo cliente":
    desconto = 0.0
elif tipo_cliente == "Cliente fidelizado":
    desconto = 0.05
elif tipo_cliente == "Cliente premium":
    desconto = 0.10
else:
    desconto = 0.0  # caso padrão para tipos desconhecidos

valor_final = valor_total * (1 - desconto)

# Exibe o resultado formatado com duas casas decimais
print(f"{valor_final:.2f}")
```

Receba uma lista de números e mostre:

o maior

o menor

a média

```python
entrada = input("Digite números separados por espaço: ")
numeros = list(map(int, entrada.split()))

if not numeros:
    print("Lista vazia. Digite ao menos um número.")
else:
    i = 0
    maior = numeros[0]
    menor = numeros[0]
    soma = 0

    while i < len(numeros):
        numero = numeros[i]

        if numero > maior:
            maior = numero
        if numero < menor:
            menor = numero

        soma += numero
        i += 1

    media = soma / len(numeros)

    print("Maior número:", maior)
    print("Menor número:", menor)
    print("Média:", media)
```

Faça um programa que mostre uma contagem regressiva de 10 até 1 e depois escreva "Feliz Ano Novo!".

```python
# Contagem regressiva de 10 até 1

contador = 10  # começamos do 10

while contador >= 1:   # enquanto contador for maior ou igual a 1
    print(contador)    # mostra o número atual
    contador -= 1      # diminui o contador em 1

print("Feliz Ano Novo!")  # mensagem final
```

Crie um programa que soma números até o usuário digitar 0

```python
# Programa que soma números até o usuário digitar 0

soma = 0  # variável para acumular a soma

numero = int(input("Digite um número (0 para parar): "))

while numero != 0:   # enquanto o número for diferente de 0
    soma += numero   # adiciona o número à soma
    numero = int(input("Digite um número (0 para parar): "))

print("Soma total:", soma)
```

# **Funções**

```python
#Exemplo
"""O principal objetivo das funções é:
Reutilizar o código: Permite escrever um bloco de código uma vez e chamá-lo várias vezes
em diferentes partes do programa, sem precisar reescrevê-lo, criando assim uma rotina para o programa.
[1]Modularidade: Ajuda a dividir um programa complexo em partes menores
 mais gerenciáveis e independentes.
[2] Isso torna o código mais fácil de entender, depurar e manter.Organização e
legibilidade: Torna o código mais limpo e lógico, dando nomes descritivos às tarefas realizadas.
Abstração: Permite que o programador use a função sem saber exatamente como ela funciona internamente
concentrando-se apenas em sua entrada (argumentos) e saída (valor de retorno)."""


def lin():
  print('-'*30)


lin()
print("Função")
lin()


def saudacao(nome):
  print("Olá",nome)

print("Chamando a função saudacao:")
saudacao("Jefferson")
saudacao("João")


#Funcao com retorno
def quadrado(numero):
  resultado = numero ** 3
  return resultado

print("Chamando a função quadrado:")
resultado = quadrado(6)
print(resultado)
```
