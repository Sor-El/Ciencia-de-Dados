```python
# dados de entrada
peso = float(input())
preco_por_tonelada = float(input())
tipo_cliente = input()

# Calculo o valor total sem desconto
valor_total = peso * preco_por_tonelada

# TODO Aplique o desconto conforme o tipo de cliente
if tipo_cliente == "Novo cliente":
    desconto = 0.0
elif tipo_cliente == "Cliente fidelizado":
    desconto = 0.05
elif tipo_cliente == "Cliente premium":
    desconto = 0.10
else:
    desconto = 0.0  # caso padrão para tipos desconhecidos

valor_final = valor_total * (1 - desconto)

# Exibe o resultado formatado com duas casas decimais
print(f"{valor_final:.2f}")
```

# Laço de Repetição em Python

```python
print("For Utilizando Lista")
Lista = [1,2,3,4,5]
for elemento in Lista:
    print(elemento)


print("For Utilizando Tupla")
tupla = (1,2,3,4,5)
for elemento in tupla:
    print(elemento)

pessoa = {"nome": "joão","idade": 30, "cidade": "Ferraz de Vasconcelos"}
print("For Utilizando Dicionário")
for chave in pessoa:
    print(chave, ":", pessoa[chave])
```

# Loop For  range();


```python
print("\n Utilizando a função range() com len()")
list = [1,2,3,4,5]
print(list)
for indice in range(0, len(list)):
    print(list[indice])

#enumerate()
lista_enumerete = ["a", "b", "c"]
for indice, valor in enumerate(lista_enumerete):
    print(indice, valor)
```

Exercício 2.5: Soma dos Números em uma Lista

```python
numeros = [10, 20, 30, 40, 50]

soma = 0
for numero in numeros:
    soma += numero

print("A soma dos números é:",soma)
```

Exercício 2.6: Encontrando o Maior Número

```python
lista_numeros = [5, 12, 3, 18, 9, 25]

maior_numero = lista_numeros[0]  # começa assumindo o primeiro elemento
for numero in lista_numeros:
    if numero > maior_numero:
        maior_numero = numero

print("O maior número da lista é:", maior_numero)
```

# **Loop** **While**

```python
print("Exemplo de while")
contador = 0
while True:
    print("Contagem",contador)
    contador += 1
    if contador == 5:
      break

print("Programa finalizado")
```

Exercício 3.1: Contador Simples
Objetivo: Contar de 1 até 5 e imprimir cada número.


```python
# Contador simples de 1 até 5
contador = 1

while contador <= 5:
    print(contador)
    contador = contador + 1  # ou contador += 1
```

# **Funções**

```python
#Exemplo
"""O principal objetivo das funções é:
Reutilizar o código: Permite escrever um bloco de código uma vez e chamá-lo várias vezes
em diferentes partes do programa, sem precisar reescrevê-lo, criando assim uma rotina para o programa.
[1]Modularidade: Ajuda a dividir um programa complexo em partes menores
 mais gerenciáveis e independentes.
[2] Isso torna o código mais fácil de entender, depurar e manter.Organização e
legibilidade: Torna o código mais limpo e lógico, dando nomes descritivos às tarefas realizadas.
Abstração: Permite que o programador use a função sem saber exatamente como ela funciona internamente
concentrando-se apenas em sua entrada (argumentos) e saída (valor de retorno)."""


def lin():
  print('-'*30)


lin()
print("Função")
lin()


def saudacao(nome):
  print("Olá",nome)

print("Chamando a função saudacao:")
saudacao("Jefferson")
saudacao("João")


#Funcao com retorno
def quadrado(numero):
  resultado = numero ** 3
  return resultado

print("Chamando a função quadrado:")
resultado = quadrado(6)
print(resultado)
```
